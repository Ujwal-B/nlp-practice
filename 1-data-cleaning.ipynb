```python
import re

def rem_single_letters(df):
    pattern = r'\s[a-zA-Z]\s'
    df['description'] = [re.sub(pattern, ' ', str(x)) for x in df['description']]

def rem_unwanted_chars(df):
    pattern = r'[\!\@\#\$\%\^\&\*\(\)\-\_\+\=\/\.\,]'
    df['description'] = [re.sub(pattern, ' ', str(x)) for x in df['description']]

def conv_to_lowercase(df):
    df['description'] = df['description'].str.lower()

def rem_extra_spaces(df):
    pattern = r'\s{2}'
    df['description'] = [re.sub(pattern, ' ', str(x)) for x in df['description']]
```

```python
def clean_text(desc):
    conv_to_lowercase(desc)
    rem_single_letters(desc)
    rem_unwanted_chars(desc)
    rem_extra_spaces(desc)
```

```python
print(desc.loc[5, 'description'])
clean_text(desc)
print(desc.loc[5, 'description'])
```
